```python
# Imports required for project
import matplotlib.pyplot as plt
import matplotlib as mlb
import pandas as pd
from ast import literal_eval
import numpy as np
import seaborn as sns
import re
```

```python
# Read csv file to dataframe
df = pd.read_csv('papers_data.csv', encoding='utf-8')
```

```python


# Capture the values of the 'published' column to array
# published= df['published'].values

"""
Capture value of year from part of pdf_url column using regex.
e.g.: pdf_url:http://arxiv.org/pdf/1701.00146v1 - get 17 from 1701.00146v1
"""
def get_year(x):
    if x!='pdf_url':
        y ='20'+re.findall('\d+',x)[0][:2]
        return y

df['year'] = df.pdf_url.apply(lambda x:get_year(x)).to_list()
# df['year'] = df.published.apply(lambda x:check(x)).to_list()


# years=[]
# for y in published:
#     try:
#         ye=literal_eval(y[0:4])
#         years.append(ye)
#     except Exception as err:
#         # print('Error :', err)
#         pass


# Show the counts of observations in each categorical bin using bars.
plot = sns.countplot(data=df, x='year', hue='primary_category')
plot.legend(title='primary_category', bbox_to_anchor=(1, 1), loc='upper left')

for cont in plot.containers:
    # set the bar labels
    plot.bar_label(cont, fmt='%.0f', label_type='edge')
```
