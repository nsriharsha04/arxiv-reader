```python
# Imports required for project
import arxiv
import json
import time
import numpy as np
from tqdm import tqdm
import pandas as pd
import csv
import warnings
warnings.filterwarnings("ignore")
```

```python
papers=[]

"""
# Function to add dictionary of arxiv result to list and make it list of dictionaries {List[Dictionary]}
"""
def addToList(result):
    
    # print(result)
    paper_info = {}
    paper_info['title'] = result.title
    paper_info['authors'] = []

    for author in result.authors:
        paper_info['authors'].append(str(author))

    paper_info['summary'] = result.summary
    paper_info['comment'] = result.comment
    paper_info['journal_ref'] = result.journal_ref
    paper_info['doi'] = result.doi
    paper_info['entry_id'] = result.entry_id
    paper_info['updated'] = str(result.updated)
    paper_info['published'] = str(result.published)
    paper_info['primary_category'] = result.primary_category
    paper_info['categories'] = result.categories
    paper_info['links'] =  str(result.links)
    paper_info['pdf_url'] = result.pdf_url

    papers.append(paper_info)

"""
# Function to write/append data to csv
"""
def writeToCsv(val : list):
    print('length of data: ',len(val))
    if len(val) != 0:
        keys = ['title','authors','summary','comment','journal_ref','doi','entry_id',\
            'updated','published','primary_category','categories','links','pdf_url']
        with open('papers_data.csv', 'a', newline='', encoding='utf-8') as output_file:
            writer = csv.DictWriter(output_file, keys)
            writer.writeheader()
            writer.writerows(val)
    else:
        print('Empty list')
```

```python

# Create array of years and months required
years  = np.arange(17,22) # from 2017 to 2021
months = np.arange(1,13) # from Jan to Dec

"""
Iterate through years [2017-2021] and for coressponding year iterate through months [1-12]
"""
for y in years:
    for m in months:
        ids_list=[]  # To hold list of paper ids.
        low = 0
        high = 500
        break_flag = False
        ctr = 0
        for k in range(1,100000):
            temp=str(y)+str(m).zfill(2)+'.'+str(k).zfill(5) # id of the format "1703.00663" is generated
            ids_list.append(temp)
            
        print('year: ',y,' month: ',m) # For tracking in logs

        # Implement Exception Handling to handle any runtime exceptions.
        try:
            while low != 100000: #Loop until it breaks or end of range is reached
            
                search = arxiv.Search(id_list=ids_list[low:high])
                result =list(search.results())

                # print('Number of results from query:',len(result))
        
                #If result has nothing but a blank , then break out of the loop (only in case of a fully empty list)
                if len(result) == 0:
                    break_flag = True #Set flag that loop was broken out of
                    break
                    
                # Going through the resultset and filtering out based on required primary_category
                for i in result:
                    if i.primary_category in ['cs.LG','cs.AI','cs.CC','cs.AR']:
                        addToList(i)
                       
                # Transfer/append the collected data to csv file before every iteration.    
                if(len(papers)!=0):
                    writeToCsv(papers)
                    papers=[]

                print('chunk: ',(low,high)) # To track the chunk size in logs
                
                print(100*"*")    # separator between loops for easy understanding of the logs. 
                # Incrementing the chunk by 500.   
                low = high #Old high becomes new low
                high += 500 #New high incremented by 500
                time.sleep(1.5) # Dose off for 1.5s before next API call, to prevent continuous arxiv call.

#  Exception handling during runtime
        except ConnectionResetError as cr:
            print(cr)    
        except AttributeError as ar:
            print(ar)          
        except Exception as err:
            print('Error', err)
           
    
```
